## Exercise 5 ‚Äì Truth Tables üêç

A **truth table** lists every possible combination of truth values for the variables in a formula and shows whether the formula is true or false for each combination.

Example for 2 variables A, B ‚Äî there are $2^2 = 4$ rows:

| A | B | A ‚àß B |
|---|---|-------|
| F | F |   F   |
| F | T |   F   |
| T | F |   F   |
| T | T |   T   |

In Python we represent a truth assignment as a **dictionary**:
```python
assignment = {'A': True, 'B': False}
```
And a formula as a **function** that takes an assignment and returns True or False:
```python
formula = lambda m: m['A'] and m['B']   # A ‚àß B
```

`itertools.product` generates all combinations for us:
```python
from itertools import product
for vals in product([False, True], repeat=2):
    print(vals)   # (False,False), (False,True), (True,False), (True,True)
```

````python
from itertools import product

def truth_table(variables, formula_fn):
    """
    Print a truth table for a formula over the given variables.
    Also returns the number of satisfying (True) rows.

    variables  : list of variable name strings, e.g. ['A', 'B']
    formula_fn : function that takes a dict {var: bool} and returns bool
    """
    # Print header
    header = " | ".join(f"{v:>5}" for v in variables) + " | result"
    print(header)
    print("-" * len(header))

    satisfying = 0

    # product([False, True], repeat=n) gives all 2^n combinations
    for vals in product([False, True], repeat=len(variables)):

        # Build assignment dict from variable names + current values
        assignment = dict(zip(variables, vals))

        # YOUR CODE HERE:
        # 1. Call formula_fn(assignment) to get the result (True or False)
        # 2. Print a row: values of each variable, then the result
        #    Hint: use  "T" if v else "F"  to convert booleans to letters
        # 3. If result is True, increment the `satisfying` counter
        
        result = formula_fn(assignment)  # replace with formula_fn(...)
        row = " | ".join(f"{'T' if v else 'F':>5}" for v in vals)
        satisfying += 1 if result else 0
        print(row + f" | {'T' if result else 'F'}")
        pass

    print(f"\nTotal satisfying models: {satisfying} / {2**len(variables)}")
    return satisfying


print("truth_table defined ‚úÖ")
````

## Exercise 6 ‚Äì Resolution üêç

**Resolution** is a proof technique for propositional logic. It works by refutation:
to prove that $KB \models \phi$, we add $\lnot\phi$ to the KB and try to derive a **contradiction** (the empty clause $\{\}$).

### Clauses and literals

In resolution, every formula must be in **CNF** ‚Äî a conjunction of **clauses**, where each clause is a **disjunction of literals**.

We represent a clause as a Python **set of strings**:
```python
{'P', '¬¨Q', 'R'}   # means  P ‚à® ¬¨Q ‚à® R
```

### The resolution rule

Given two clauses that contain **complementary literals** (e.g. `P` and `¬¨P`),  
we can derive a new clause ‚Äî the **resolvent** ‚Äî by combining both clauses minus those complementary literals:

```
C1 = {P, A, B}     C2 = {¬¨P, C, D}
resolvent = {A, B, C, D}   (remove P and ¬¨P, merge the rest)
```

If the resolvent is the **empty set** `{}`, we found a contradiction ‚Üí proof complete!

````python
def negate(literal):
    """
    Returns the negation of a literal.
    negate('P')  -> '¬¨P'
    negate('¬¨P') -> 'P'
    """
    if literal.startswith('¬¨'):
        return literal[1:]          # strip the ¬¨
    else:
        return '¬¨' + literal        # prepend ¬¨


def resolve(c1, c2):
    """
    Try to resolve two clauses (given as sets of literal strings).
    Returns the resolvent set if resolution is possible, else None.

    Example:
        resolve({'P', 'A'}, {'¬¨P', 'B'})  ->  {'A', 'B'}
        resolve({'P', 'Q'}, {'A', 'B'})   ->  None  (no complementary literals)
    """
    for lit in c1:
        if negate(lit) in c2:
            # YOUR CODE HERE:
            # Build the resolvent:
            #   - take everything from c1 EXCEPT lit
            #   - union with everything from c2 EXCEPT negate(lit)
            # Hint: set difference:  c1 - {lit}
            # Hint: set union:       set_a | set_b
            resolvent = (c1 - {lit}) | (c2 - {negate(lit)})
            return resolvent
    return None   # no complementary literals found


# ‚îÄ‚îÄ Quick test ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
r = resolve({'P', 'A'}, {'¬¨P', 'B'})
print(f"resolve({{'P','A'}}, {{'¬¨P','B'}}) = {r}")
print(f"Expected: {{'A', 'B'}}")
print()
r2 = resolve({'P', 'Q'}, {'A', 'B'})
print(f"resolve({{'P','Q'}}, {{'A','B'}}) = {r2}")
print(f"Expected: None")
````

````python
def pl_resolution(kb_clauses, goal_literals):
    """
    Attempt to prove a goal from KB using resolution by refutation.

    kb_clauses    : list of clauses (each clause is a set of literal strings)
    goal_literals : set of literals representing the goal (e.g. {'Magical'})

    How it works:
      1. Negate every literal in goal_literals and add each as a unit clause.
      2. Repeatedly try to resolve pairs of clauses.
      3. If we derive the empty clause {} ‚Üí contradiction ‚Üí goal is PROVED.
      4. If no new clauses can be derived ‚Üí goal CANNOT be proved.
    """
    # Step 1: negate the goal and add as unit clauses
    negated_goal = [frozenset({negate(lit)}) for lit in goal_literals]

    # Work with frozensets so we can put clauses in a set
    clauses = set(frozenset(c) for c in kb_clauses) | set(negated_goal)
    new_clauses = set()

    iteration = 0
    while True:
        iteration += 1
        clause_list = list(clauses)

        # Try every pair of clauses
        for i in range(len(clause_list)):
            for j in range(i + 1, len(clause_list)):
                c1 = set(clause_list[i])
                c2 = set(clause_list[j])

                # YOUR CODE HERE:
                # 1. Call resolve(c1, c2) to get the resolvent
                # 2. If the resolvent is the EMPTY SET ‚Üí return True (proved!)
                # 3. Otherwise if resolvent is not None ‚Üí add it to new_clauses
                resolvent = resolve(c1, c2)
                if resolvent is not None:
                    if len(resolvent) == 0:
                        return True
                    else:
                        new_clauses.add(frozenset(resolvent))

        # If no new clauses were found, we cannot prove the goal
        if new_clauses.issubset(clauses):
            return False

        clauses |= new_clauses


print("pl_resolution defined ‚úÖ")
````

## Exercise 7 ‚Äì First-Order Logic: Formula Evaluator üêç

FOL formulas quantify over a **domain** of objects.  
We can check FOL formulas computationally by iterating over a finite domain.

We will use a small domain of animals: cats and dogs, big and small.

A **model** in FOL is:
- A domain (set of objects)
- An interpretation of each predicate (which objects satisfy it)

```python
domain = ['cat1', 'cat2', 'dog1']
big    = {'cat1', 'dog1'}      # big(x) is true for these objects
cat    = {'cat1', 'cat2'}      # cat(x) is true for these objects
```

````python
# ‚îÄ‚îÄ Domain setup ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
domain = ['cat1', 'cat2', 'dog1', 'dog2']
big    = {'cat1', 'dog1', 'dog2'}   # big(x)
cat    = {'cat1', 'cat2'}           # cat(x)


# ‚îÄ‚îÄ Helper: check a formula for all x in domain ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ

def forall(domain, predicate):
    """Returns True if predicate(x) is True for EVERY x in domain."""
    return all(predicate(x) for x in domain)

def exists(domain, predicate):
    """Returns True if predicate(x) is True for AT LEAST ONE x in domain."""
    return any(predicate(x) for x in domain)


# ‚îÄ‚îÄ 7.1: Evaluate the 7 formulas from the worksheet ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Formula 1: ‚àÉx. (big(x) ‚àß cat(x))  ‚Üí  "There exists a big cat"
f1 = exists(domain, lambda x: x in big and x in cat)
print(f"Formula 1 ‚Äî ‚àÉx.(big‚àßcat):        {f1}   ‚Üê expected True (cat1 is big and a cat)")

# Formula 4: ‚àÄx. (big(x) ‚àß cat(x))  ‚Üí  "All things are big cats"
f4 = forall(domain, lambda x: x in big and x in cat)
print(f"Formula 4 ‚Äî ‚àÄx.(big‚àßcat):        {f4}  ‚Üê expected False (dog2 is not a cat)")

# Formula 5: ‚àÄx. (big(x) ‚Üí cat(x))  ‚Üí  "All big things are cats"
f5 = forall(domain, lambda x: x not in big or x in cat)
print(f"Formula 5 ‚Äî ‚àÄx.(big‚Üícat):        {f5}  ‚Üê expected False (dog1 is big but not a cat)")

print()

# YOUR CODE HERE ‚Äî evaluate formulas 2, 3, 6, 7:
# Formula 2: ‚àÉx. (big(x) ‚Üí cat(x))   (remember: A‚ÜíB = ¬¨A ‚à® B)
# Formula 3: ‚àÉx. (big(x) ‚à® cat(x))
# Formula 6: ‚àÄx. (big(x) ‚à® cat(x))
# Formula 7: ‚àÄx. (big(x) ‚à® ¬¨cat(x))
pass
````
